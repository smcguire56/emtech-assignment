```python
# Import pandas.
import pandas as pd
```

```python
# Load the iris data set from a URL.
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df
```

```python
import seaborn as sns
```

```python
df.head()
```

```python
a=sns.pairplot(df,hue='species',palette="husl",size=5,vars=['sepal_width','sepal_length','petal_length','petal_width'],kind='scatter',markers=['o','x','+'])

# src: https://seaborn.pydata.org/generated/seaborn.pairplot.html
```
