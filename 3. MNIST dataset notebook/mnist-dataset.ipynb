```python
# get the labels
import array as arr
import os



def mkDir(folderName):
    if not os.path.exists(folderName):
        os.makedirs(folderName) 
mkDir('0')
mkDir('1')
mkDir('2')
mkDir('3')
mkDir('4')
mkDir('5')
mkDir('6')
mkDir('7')
mkDir('8')
mkDir('9')

myLabels = []
for x in range(8, 10000, 1):
    myLabels += [int.from_bytes(labels[x:x+1], byteorder="big")]

#for x in myLabels:
#    print(x)

#for i in range(0, 10008):
#    print(myLabels[i])
```
